## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

### I need more infomation about the mjtheme_namecode column before proceeding.

```python
worldbank_json_df['mjtheme_namecode'].head()
```

### This looks like a nested dictionary, and I will need a new approach to read the data cleanly.

```python
data = json.load(open('data/world_bank_projects.json')) #this reads in an array, which allows use of json_normalize, below
data[0:2]
```

```python
code = json_normalize(data, 'mjtheme_namecode',['id'])
code.head()
```

```python
topten_codes = code.code.value_counts().head(10)
topten_codes
```

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

### My approach is to create a key/value table or dictionary and then perform a join/merge.

```python
import numpy as np
UniqueCodes = np.unique(code['code'])
print(UniqueCodes)
print(len(UniqueCodes))
```

```python
UniqueLabels = np.unique(code['name'])
print(UniqueLabels) # this includes the 'blank' labels
print(len(UniqueLabels)) # one longer than the Code array
```

```python
print(UniqueLabels[1:])
len(UniqueLabels[1:]) # use slicing to eliminate the blank entry 
```

```python
UniqueLabsFixed = UniqueLabels[1:]
len(UniqueLabsFixed) #save new array without the blank
```

### This approach has lost the unique pairing, and I need to use a different strategy.  drop_duplicates will be more useful here.

```python
unique = code[['code','name']].drop_duplicates() #this is tidier
unique
```

```python
unique = unique[unique.name != ''] # remove the blanks
unique
```

```python
unique = unique.reset_index().drop('index',axis=1) # reset index
unique
```

### Now let's merge to the old dataframe, code, and get rid of extraneous columns.

```python
merge = code.merge(unique, how='left',  on='code', suffixes = ('_old','')).drop(['name_old','id' ], axis=1)
merge
```

```python
merge.describe() # check number of unique values
```
